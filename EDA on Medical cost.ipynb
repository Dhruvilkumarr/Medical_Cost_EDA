## Univariate Analysis

In this section, we will perform univariate analysis to explore the distribution of individual features in the dataset. We will use histograms with kernel density estimates (KDE) to visualize the distributions.

1. **Distribution of Age**: We will plot the distribution of ages to understand the age range and frequency of occurrence within the dataset.

2. **Distribution of BMI**: We will plot the distribution of Body Mass Index (BMI) to assess the range and frequency of BMI values.

3. **Distribution of Medical Costs**: We will plot the distribution of medical costs to visualize the cost range and how frequently different cost levels occur.


```python
#age distribution
plt.figure(figsize=(5, 3)) 
AgePlt = sns.boxplot(data=df, x='Age')
plt.legend(['Age Distribution'], loc='upper right')
AgePlt.set_xticks(np.arange(15, 80, 5))
AgePlt.set_xlabel('Age')
plt.show()

#BMI distribution
plt.figure(figsize=(7, 3)) 
BMIPlt = sns.boxplot(data=df, x='BMI')
plt.legend(['BMI Distribution'], loc='upper right')
BMIPlt.set_xticks(np.arange(15, 40, 1))
BMIPlt.set_xlabel('BMI')
plt.show()

#Children distribution
plt.figure(figsize=(7, 3)) 
KidsPlt = sns.boxplot(data=df, x='Children')
plt.legend(['Kids Distribution'], loc='upper right')
KidsPlt.set_xlabel('Kids')
plt.show()

#Medical Cost distribution
plt.figure(figsize=(7, 3)) 
MPlt = sns.boxplot(data=df, x='Medical Cost')
plt.legend(['Cost Distribution'], loc='upper right')
MPlt.set_xlabel('Cost')
plt.show()
```

## Bivariate Analysis - Box Plots

In this section, we will perform bivariate analysis using box plots to explore the relationship between medical costs and categorical variables.

1. **Medical Cost by Sex**: We will create a box plot to examine how medical costs vary between different sexes.

2. **Medical Cost by Smoker Status**: We will create a box plot to analyze the distribution of medical costs for smokers versus non-smokers.

3. **Medical Cost by Region**: We will create a box plot to observe how medical costs differ across various regions.


```python

#Medical Cost by Sex
sns.boxplot(data=df, y='Sex', x='Medical Cost')
plt.show()

#Medical Cost by Smoker Status
sns.boxplot(data=df, y='Smoker', x='Medical Cost')
plt.show()

#Medical Cost by Region
sns.boxplot(data=df, y='Region', x='Medical Cost')
plt.show()
```

## Analysis of Smokers

In this section, we analyze smokers based on gender and age using bar and line charts.

1. **Count of Smokers by Gender**: We will create a bar plot to visualize the number of smokers categorized by gender. This will help us understand the distribution of smokers across different genders.

2. **Number of Smokers by Age**: We will create a line chart to show how the number of smokers varies with age. This will provide insight into the age distribution of smokers within the dataset.


```python
# Smokers by Gender

# Filter data for smokers

Smoker = df[df['Smoker']== 'yes']
# Count the number of smokers by gender
Smoker_Count = Smoker['Sex'].value_counts().reset_index()
# Plotting the count of smokers by gender
plt.figure(figsize=(5, 7)) 
CountPlt = sns.barplot(data=Smoker_Count, x='Sex', y='count')
plt.legend(['Smokers'], loc='upper right')
CountPlt.set_yticks(np.arange(0, 2600, 250))
plt.show()

#Line Chart for Smokers by Age

# Count the number of smokers by age
SmkAge = Smoker['Age'].value_counts().reset_index()
# Plotting the number of smokers by age
plt.figure(figsize=(10, 3)) 
CountAgePlt = sns.lineplot(data=SmkAge, x='Age', y='count')
plt.legend(['Smokers'], loc='upper right')
CountAgePlt.set_xticks(np.arange(16, 70, 2))
plt.show()
```

## Multivariate Analysis

In this section, we will analyze the relationship between multiple variables. 

- **Comparison of Medical Costs by BMI for Smokers and Non-Smokers**: We will use a line plot to compare medical costs based on BMI, with separate lines for smokers and non-smokers. This visualization helps us understand how medical costs vary with BMI and how this relationship differs between smokers and non-smokers.


```python
# Multivariate Analysis

# Plotting the comparison of medical costs by BMI for smokers and non-smokers
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='BMI', y='Medical Cost', hue='Smoker', marker='o')
plt.title('Comparison of Medical Costs by Age for Smokers and Non-Smokers')
plt.xlabel('Age')
plt.ylabel('Medical Cost')
plt.legend(title='Smoker')
plt.show()
```
